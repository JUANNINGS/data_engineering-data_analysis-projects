```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('IMBA_DS').getOrCreate()
```

### Prepare data sources

```
data = spark.read.csv('/FileStore/tables/data_small.csv', header=True, inferSchema=True)
display(data)
```

```
orders = spark.read.csv('/FileStore/tables/orders.csv', header=True, inferSchema=True)
display(orders.head(10))
```

```
order_product_train = spark.read.csv('/FileStore/tables/order_products__train.csv', header=True, inferSchema=True)
display(order_product_train.head(10))
```

### More data preprocessing

```
# attach user_id to order_product_train
order_product_train = order_product_train.join(orders[['user_id','order_id']], on=orders.order_id == order_product_train.order_id).drop(order_product_train['order_id'])

display(order_product_train.head(10))
```

```
# attach eval_set to data
data = data.join(orders[orders.eval_set != 'prior'][['user_id','eval_set']], on=data.user_id == orders.user_id).drop(orders['user_id'])
display(data)
```

```
order_product_train = order_product_train.withColumnRenamed('user_id', 'opt_user_id')
```

```
# attach target variable: reordered
data = data.join(order_product_train[['opt_user_id', 'product_id','reordered']], on=(data.user_id == order_product_train.opt_user_id) & (data.product_id == order_product_train.product_id), how = 'left').drop(order_product_train['product_id']).drop('opt_user_id')

display(data)
```

```
# a few more feature engineering
data = data.withColumn('prod_reorder_probability', data.prod_second_orders / data.prod_first_orders)
data = data.withColumn('prod_reorder_times', 1 + data.prod_reorders / data.prod_first_orders)
data = data.withColumn('prod_reorder_ratio', data.prod_reorders / data.prod_orders)
data = data.withColumn('user_average_basket', data.user_total_products / data.user_orders)
data = data.withColumn('up_order_rate', data.up_orders / data.user_orders)
data = data.withColumn('up_orders_since_last_order', data.user_orders - data.up_last_order)
data = data.withColumn('up_order_rate_since_first_order', data.up_orders / (data.user_orders - data.up_first_order + 1))
```

```
data = data.drop('prod_reorders', 'prod_first_orders', 'prod_second_orders')
```

```
display(data)
```

```
# split into training and test set, test set does not have target variable
train = data[data.eval_set == 'train']
test = data[data.eval_set == 'test']
display(train)
```

```
# convert target variable to 1/0 for training dataframe
train = train.na.fill(0, ['reordered'])
```

```
# drop id columns as they won't be used in model
train = train.drop('eval_set', 'user_id', 'product_id')
display(train)
```

###  Prepare data features and data labels for Machine Learning

```
from pyspark.ml.feature import VectorAssembler
```

```
assembler = VectorAssembler(
  inputCols=['up_orders',
 'user_mean_days_since_prior',
 'user_period',
 'user_distinct_products',
 'user_reorder_ratio',
 'user_total_products',
 'up_average_cart_position',
 'up_first_order',
 'user_orders',
 'up_last_order',
 'prod_orders',
 'prod_reorder_probability',
 'prod_reorder_times',
 'prod_reorder_ratio',
 'user_average_basket',
 'up_order_rate',
 'up_orders_since_last_order',
 'up_order_rate_since_first_order'],
              outputCol="features")

output = assembler.transform(train)
display(output)
```

```
final_data = output.select('features','reordered')
display(final_data.head(5))
```

```
train_data,val_data = final_data.randomSplit([0.7,0.3])
display(train_data.count())
```

### Use Random Forest to build the machine learning model

```
from pyspark.ml.classification import RandomForestClassifier
```

```
rfc = RandomForestClassifier(labelCol='reordered',featuresCol='features', maxDepth=5, numTrees=100, seed=1)
rfc_model = rfc.fit(train_data)
rfc_predictions = rfc_model.transform(val_data)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator

auc_evaluator = BinaryClassificationEvaluator(labelCol="reordered", metricName='areaUnderROC')
rfc_auc = auc_evaluator.evaluate(rfc_predictions)
display(rfc_auc)
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

acc_evaluator = MulticlassClassificationEvaluator(labelCol="reordered", metricName='accuracy')
rfc_acc = acc_evaluator.evaluate(rfc_predictions)
display(rfc_acc)
```
